# Function to Find Data for Each Team per Year

```
def getSeasonData(team, year):
  team_data = df_nfl.loc[team, year]
  ypp = team_data['yards/play']
  pypa = team_data['net yards per pass att']
  rypa = team_data['rush yards per att']
  sp = team_data['score percentage']
  top = team_data['turnover percentage']
  appd = team_data['avg points per drive']
  oypp = team_data['opp yards/play']
  opypa = team_data['opp net yards per pass att']
  orypa = team_data['opp rush yards per att']
  osp = team_data['opp score percentage']
  otop = team_data['opp turnover percentage']
  oappd = team_data['opp avg points per drive']
  return [ypp, pypa, rypa, sp, top, appd, oypp, opypa, orypa, osp, otop, oappd]
```

```
def compareTeams(team_1, team_2, year):
  t1 = np.array(getSeasonData(team_1, year))
  t2 = np.array(getSeasonData(team_2, year))
  diff = t1 - t2
  return diff
```

```
def createDict(year):
  d = {}
  df2 = df_nfl[df_nfl['Year'] == year]
  for x in range(len(df2)):
    d[df2.iloc[x]['Team']] = getSeasonData(df2.iloc[x]['Team'], year)
  return d
```

```
def getHome(at):
  if at == '@':
    return -1
  elif at == ' ':
    return 1
  else:
    return 0
```

```
def createSets(years):
  totalGames = len(df)
  xTrain = np.zeros((totalGames, 13))
  yTrain = np.zeros((totalGames))
  i = 0
  for year in years:
    season_dict = createDict(str(year))
    #print(season_dict)
    season = df[df['Year'] == year]
    xTrainSeason = np.zeros((df[df['Year'] == year]['Year'].count(), 13))
    yTrainSeason = np.zeros((df[df['Year'] == year]['Year'].count()))
    count = 0
    #print(xTrainSeason)
    
    for x in range(len(season)):
      wTeam = season.iloc[x]['Winning Team']
      wStats = wTeam.split()
      mascot = wStats[len(wStats)-1]
      wStats = season_dict[mascot]
      lTeam = season.iloc[x]['Losing Team']
      lStats = lTeam.split()
      mascot2 = lStats[len(lStats)-1]
      lStats = season_dict[mascot2]
      home = getHome(season.iloc[x]['At?'])
      diff = compareTeams(mascot, mascot2, str(year))
      if count % 2 == 0:
        diff = np.append(diff, home)
        xTrainSeason[count] = diff
        yTrainSeason[count] = 1
      else:
        diff = np.append(diff, -home)
        xTrainSeason[count] = -diff
        yTrainSeason[count] = 0
      count += 1
    #print(yTrainSeason)
    xTrain[i:(df[df['Year'] == year]['Year'].count()) + i] = xTrainSeason
    yTrain[i:(df[df['Year'] == year]['Year'].count()) + i] = yTrainSeason
    i += df[df['Year'] == year]['Year'].count()
  return xTrain, yTrain
  
```

```
X_train, Y_train = createSets(np.arange(1999, 2020))
```

```
X_train.shape
```

```
x_train, x_test, y_train, y_test = train_test_split(X_train, Y_train, test_size=0.2)
x_train
```

# Recurrent Neural Network to Predict Which Team Will Win

```
model = models.Sequential()
model.add(layers.Dense(8, activation='relu', input_shape=(13,)))
model.add(layers.Dense(8, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=3)
```

```
results = model.evaluate(x_test, y_test)
```
